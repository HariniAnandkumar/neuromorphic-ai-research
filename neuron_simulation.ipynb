```python
membrane_potential = 0.0
threshold = 1.0
leak = 0.1
refractory_steps = 2
refractory_timer = 0

inputs = [0.2, 0.3, 0.1, 0.4, 0.2, 0.3, 0.4]

for t in range(len(inputs)):
    if refractory_timer > 0:
        print(f"Time {t} | Refractory")
        refractory_timer -= 1
        continue

    membrane_potential *= (1 - leak)
    membrane_potential += inputs[t]

    print(f"Time {t} | V = {membrane_potential:.2f}")

    if membrane_potential >= threshold:
        print("Spike :)")
        membrane_potential = 0.0
        refractory_timer = refractory_steps
```
